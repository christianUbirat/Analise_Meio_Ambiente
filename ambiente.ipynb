Primeiramente, vale a pena dedicar um tempo para entender a quanto tempo o homem existe em relação à Terra e ao Universo.

Estimasse que o Big Bang ocorreu há cerca de 13,8 bilhões de anos (ou 13.800.000.000 anos). Já o planeta Terra, há cerca de 4,5 bilhões de anos.  As formas mais simples de vida entre 4 e 3,5 bilhões de anos. E os primeiros Homo Sapiens, cerca de 200.000 anos. 
Esses números por si só parecem pura abstração, então para entender melhor, vamos imaginar uma corda que vai da Terra até o Sol (ou seja, aproximadamente 150 milhões quilômetros). 
Se imaginarmos que essa corda representa a linha do tempo do universo desde o Big Bang, com o fim da corda representando o presente, então cada 11 metros reprentaria aproximadamente 1 ano. 
Nessa representação o surgimento do planeta Terra distaria aproximadamente 409.090km do fim de nossa corda, o suficiente para vir da Lua até nosso planeta e ainda dar quase duas voltas na Terra. 
O surgimento da primeira forma de vida distaria aproximadamente 318.181km da ponta que representa o presente, o que é suficiente para dar quase 8 voltas ao redor da Terra.
Eis agora que iremos do presente até o ponto da corda que representa o surgimento do primeiro Homo Sapiens. Para isso, bastava andarmos 18km! Sequer precisaríamos caminhar meia maratona. 
Visualmente temos:

```python
import numpy as np
import matplotlib.pyplot as plt
data = {'Big Bang':1254545,'Terra':409090, 'Vida':318181, 'Homo Sapiens':18}

f, ax = plt.subplots(figsize=(20,10))
plt.bar( data.keys(), data.values())
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.ylabel("Distância até o fim da corda", fontsize = 20)
plt.xlabel("Surgimentos", fontsize=20)
plt.savefig('Dist.png', format='png')
```

```python
data = {'Big Bang':1254545,'Terra':409090, 'Vida':318181, 'Homo Sapiens x 1000':18*1000}

f, ax = plt.subplots(figsize=(20,10))
plt.bar( data.keys(), data.values())
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.suptitle('Distância até o fim da corda multiplicando a distância do Homo Sapiens por 1000', fontsize=20)
plt.ylabel("Distância até o fim da corda", fontsize = 20)
plt.xlabel("Surgimentos", fontsize=20)
plt.savefig('Dist_Homem.png', format='png')
```

```python
import pandas as pd

df = pd.read_csv('./estudo.csv', delimiter=';', skiprows=0, low_memory=False)
cols = ['date', 'max']
df16 = pd.DataFrame(columns = cols)
df17 = pd.DataFrame(columns = cols)
df18 = pd.DataFrame(columns = cols)
df19 = pd.DataFrame(columns = cols)
df20 = pd.DataFrame(columns = cols)
for i in range(0, df.shape[0]):
    if ((df['date'][i])[-4:] == '2016'):
        df16.loc[i, :] = df.loc[i,:]
for i in range(0, df.shape[0]):
    if ((df['date'][i])[-4:] == '2017'):
        df17.loc[i, :] = df.loc[i,:]
for i in range(0, df.shape[0]):
    if ((df['date'][i])[-4:] == '2018'):
        df18.loc[i, :] = df.loc[i,:]
for i in range(0, df.shape[0]):
    if ((df['date'][i])[-4:] == '2019'):
        df19.loc[i, :] = df.loc[i,:]
for i in range(0, df.shape[0]):
    if ((df['date'][i])[-4:] == '2016'):
        df20.loc[i, :] = df.loc[i,:]
```

```python
anos = {'2016':df16['max'].mean(), '2017':df17['max'].mean(), '2018':df18['max'].mean(), '2019':df19['max'].mean(), '2020':df20['max'].mean()}

f, ax = plt.subplots(figsize=(20,10))
plt.plot(list(anos.keys()), list(anos.values()))
plt.xticks(fontsize= 20)
plt.yticks(fontsize= 15)
plt.suptitle('Air Quality Index geral medido pela estação em Itaim Paulista ', fontsize=30)
plt.ylabel("Air Quality Index", fontsize = 30)
plt.xlabel("Ano", fontsize=30)
plt.savefig('AQI_Itaim.png', format='png')
```

```python
oz = df = pd.read_csv('./ozonio.csv', delimiter=';', skiprows=0, low_memory=False)

f, ax = plt.subplots(figsize=(20,10))
plt.plot(oz['year'],oz['avg_hole'])
plt.xticks(fontsize= 20)
plt.yticks(fontsize= 15)
plt.suptitle('Tamanho médio do buraco na camada de ozônio no Hemisfério Sul', fontsize=30)
plt.ylabel("Tamanho (milhões km2)", fontsize = 30)
plt.xlabel("Ano", fontsize=30)
plt.savefig('ozonio.png', format='png')
```

```python
co2 = df = pd.read_csv('./co2.csv', delimiter=';', skiprows=0, low_memory=False)

f, ax = plt.subplots(figsize=(20,10))
plt.plot(co2['date'],co2['average'])
plt.xticks(fontsize= 20)
plt.yticks(fontsize= 15)
plt.suptitle('Nível de CO2', fontsize=30)
plt.ylabel("CO2 (partes por milhão)", fontsize = 30)
plt.xlabel("Ano", fontsize=30)
plt.savefig('co2.png', format='png')
```
